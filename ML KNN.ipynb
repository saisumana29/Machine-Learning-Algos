```python
import numpy as np
import pandas as pd

# Split data into training and testing sets
split_ratio = 0.8
split_index = int(len(df) * split_ratio)
df_train, df_test = df[:split_index], df[split_index:]

# Predefined parameters
distance_metric = 'C'  # Predefined as Cartesian
distance_order = 2    # Predefined as 2
k_neighbors = 5       # Increase k for better accuracy (you can tune this)

# Function to calculate Cartesian distance
def calculate_cartesian_distance(x1, x2):
    return np.linalg.norm(x1 - x2)

# Function to calculate Manhattan distance
def calculate_manhattan_distance(x1, x2):
    return np.sum(np.abs(x1 - x2))

# Function to calculate Minkowski distance
def calculate_minkowski_distance(x1, x2, order):
    return np.power(np.sum(np.power(np.abs(x1 - x2), order)), 1/order)

# Function to calculate accuracy
def calculate_accuracy(test_data, predictions):
    correct_predictions = sum(predictions == test_data['Gender'])
    return correct_predictions / len(test_data)

# Function to make predictions using k-NN
def k_nearest_neighbors_predict(training_data, x_test, k, distance_metric, distance_order):
    distances = []

    for index, row in training_data.iterrows():
        train_data = row[['Height', 'Weight', 'age']].values
        if distance_metric == 'C':  # Cartesian
            dist = calculate_cartesian_distance(x_test, train_data)
        elif distance_metric == 'M':  # Manhattan
            dist = calculate_manhattan_distance(x_test, train_data)
        elif distance_metric == 'K':  # Minkowski
            dist = calculate_minkowski_distance(x_test, train_data, distance_order)

        distances.append((dist, row['Gender']))

    distances.sort(key=lambda x: x[0])
    k_nearest = distances[:k]
    k_nearest_labels = [item[1] for item in k_nearest]

    # Count the most common label in k-nearest neighbors
    prediction = max(set(k_nearest_labels), key=k_nearest_labels.count)

    return prediction

# Function to compute confusion matrix
def compute_confusion_matrix(test_data, predictions):
    unique_labels = test_data['Gender'].unique()
    matrix = pd.DataFrame(0, index=unique_labels, columns=unique_labels)

    for true_label, predicted_label in zip(test_data['Gender'], predictions):
        matrix.loc[true_label, predicted_label] += 1

    return matrix

# Function to predict gender for a single input record
def predict_gender_single_input(input_data, training_data, k, distance_metric, distance_order):
    input_data = np.array(input_data)
    prediction = k_nearest_neighbors_predict(training_data, input_data, k, distance_metric, distance_order)
    return prediction

# Main code for testing
predictions = []

for index, row in df_test.iterrows():
    x_test = row[['Height', 'Weight', 'age']].values
    true_label = row['Gender']
    predicted_label = k_nearest_neighbors_predict(df_train, x_test, k_neighbors, distance_metric, distance_order)
    predictions.append(predicted_label)

accuracy = calculate_accuracy(df_test, predictions)
confusion = compute_confusion_matrix(df_test, predictions)

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(confusion)

# Example of predicting gender for a single input record
input_record = [170, 70, 30]  # Replace with your input values
predicted_gender = predict_gender_single_input(input_record, df_train, k_neighbors, distance_metric, distance_order)
print(f"Predicted Gender: {predicted_gender}")
```
